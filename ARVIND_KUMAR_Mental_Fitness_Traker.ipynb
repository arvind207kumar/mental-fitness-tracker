#Encoding the Categorical **values**

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in df.columns:
    if df[i].dtype == 'object':
        df[i]=l.fit_transform(df[i])
```

#MODELLING

####Assigning feautre of Matrix and Dependent Variable

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

####Spliting the Dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# how the data is split into the different group
print('How split: \n')
print('X train',x_train.shape)
print('X test',x_test.shape)
print('y train',y_train.shape)
print('y test',x_test.shape)
```

#STANDERDIZE

```python
scaler = StandardScaler()
x_train1 = pd.DataFrame(scaler.fit_transform(x_train),columns = x_train.columns)
```

#1)Linear Regression Model





```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python

lr.fit(x_train,y_train)

#predicting the value

y_pred=lr.predict(x_test)

#evaluating the model

from sklearn.metrics import r2_score,mean_squared_error
print("Results for Linear Regression:\n1)Mean Square Error={}\n2)R-Square Score={}".format(mean_squared_error(y_test,y_pred),r2_score(y_test,y_pred)))


```

```python
print("Accuracy of mode:  ",lr.score(x_test,y_test))
```

 cross validated score to ensure

```python
csv = cross_val_score(lr,x_train,y_train,cv=10)
print('accuracy of model after csv: ',np.mean(csv))
```

#2)SVM Regression Model

```python
svr=SVR()
svr.fit(x_train,y_train)

#predicting the value

y_pred=svr.predict(x_test)

#evaluating the model

from sklearn.metrics import r2_score,mean_squared_error
print("Results for SVM Regression:\n1)Mean Square Error={}\n2)R-Square Score={}".format(mean_squared_error(y_test,y_pred),r2_score(y_test,y_pred)))


```

```python
print("Accuracy of mode:  ",svr.score(x_test,y_test))
```

cross validated score to ensure

```python
csv = cross_val_score(svr,x_train,y_train,cv=10)
print('accuracy of model after csv: ',np.mean(csv))
```

#3)Decision Tree Regression Model


```python
dtr=DecisionTreeRegressor(random_state=0)
dtr.fit(x_train,y_train)

#predicting the value

y_pred=dtr.predict(x_test)

#evaluating the model

from sklearn.metrics import r2_score,mean_squared_error
print("Results for Decision Tree Regression:\n1)Mean Square Error={}\n2)R-Square Score={}".format(mean_squared_error(y_test,y_pred),r2_score(y_test,y_pred)))

```

```python
print("Accuracy of mode:  ",dtr.score(x_test,y_test))
```

```python
csv = cross_val_score(dtr,x_train,y_train,cv=10)
print('accuracy of model after csv: ',np.mean(csv))
```

#4)Random forest Regression Model

```python
rfr= RandomForestRegressor(n_estimators=10,random_state=42)
rfr.fit(x_train,y_train)

#predicting the value

y_pred=rfr.predict(x_test)

#evaluating the model

#from sklearn.metrics import r2_score,mean_squared_error
print("Results for Decision Tree Regression:\n1)Mean Square Error={}\n2)R-Square Score={}".format(mean_squared_error(y_test,y_pred),r2_score(y_test,y_pred)))


```

```python
print("Accuracy of mode:  ",rfr.score(x_test,y_test))
```

```python
csv = cross_val_score(rfr,x_train,y_train,cv=10)
print('accuracy of model after csv: ',np.mean(csv))
```
